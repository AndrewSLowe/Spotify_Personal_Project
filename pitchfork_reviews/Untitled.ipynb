```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("./database.sqlite")
```

```python
# Create a cursor object

cur = conn.cursor()
```

```python
cur.execute("select * from content;")
```

```python
df = pd.read_sql_query("SELECT * FROM content", conn)
df.head()
```

```python
df_genres = pd.read_sql_query("SELECT * FROM genres", conn)
df_labels = pd.read_sql_query("SELECT * FROM labels", conn)
df_reviews = pd.read_sql_query("SELECT * FROM reviews", conn)
df_years = pd.read_sql_query("SELECT * FROM years", conn)

tables = [df_genres, df_labels, df_reviews, df_years]
```

```python
for i in tables:
    print(i.columns)
```

```python
for i in tables:
    df = pd.merge(df, i, on='reviewid')
```

```python
print(df.shape)
df.head()
```

```python
df_dropped_dups = df[df['reviewid'].duplicated() == False]
```

```python
print(df_dropped_dups.shape)
df_dropped_dups.head()
```
